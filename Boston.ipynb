```python
X = df[["CRIM","ZN","INDUS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT"]]
y = df["MEDV"]

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 5)
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test) 
y_pred
```

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1.head()
```

```python
from sklearn import metrics 
print('R2- SCORE:', metrics.r2_score(y_test,y_pred))
```

**Lasso Regression**

```python
from sklearn.linear_model import Lasso
```

```python
Lasso_Reg = Lasso()
Lasso_Reg.fit(X_train, y_train)
```

```python
y_pred1 = Lasso_Reg.predict(X_test)
y_pred1
```

```python
df2 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred1})
df2.head()
```

```python
from sklearn import metrics 
print('R2- SCORE:', metrics.r2_score(y_test,y_pred1))
```

**Ridge Regression**

```python
from sklearn.linear_model import Ridge
Ridge_Reg = Ridge()
Ridge_Reg.fit(X_train, y_train)
```

```python
y_pred2 = Ridge_Reg.predict(X_test) # given the features of the test dataset, it will give the final predictions
"Syntax : varName.predict(test_features)"
y_pred2
```

```python
df3 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred2})
df3.head()
```

```python
from sklearn import metrics 
print('R2- SCORE:', metrics.r2_score(y_test,y_pred2))
```

**Cross Validation**

```python
from sklearn.model_selection import cross_validate
regressor = LinearRegression(normalize = True) #untrained model
cv_results = cross_validate(regressor, X, y, cv=10, scoring = "r2")
cv_results['test_score'].mean()
```
